```python
import numpy as np
import pandas as pd
```

```python
books =pd.read_csv('/content/Books.csv')
users=pd.read_csv('/content/Users.csv')
ratings= pd.read_csv('/content/Ratings.csv')
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

```python
ratings_count_book=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_count_book.rename(columns={'Book-Rating':'count-Rating'},inplace=True)
ratings_count_book.head()
ratings_count_book.shape
```

```python
avg_ratings=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating':'avg-Rating'},inplace=True)
avg_ratings.head()
avg_ratings.shape
```

```python
popularity_df=ratings_count_book.merge(avg_ratings,on='Book-Title')
popularity_df
```

```python
popularity_df=popularity_df[popularity_df['count-Rating']>=120].sort_values('avg-Rating',ascending=False).head(50)
```

```python
popularity_df.head()
```

```python
popularity_df=popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','ISBN','Book-Author','count-Rating','Year-Of-Publication','Publisher','Image-URL-M','avg-Rating']]
```

```python
popularity_df
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
jagruk_users=x[x].index
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(jagruk_users)]
filtered_ratings
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=30
famous_books=y[y].index
# famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings=final_ratings.drop_duplicates()
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_score=cosine_similarity(pt)
cosine_score
```

```python
cosine_score.shape
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]

    similar_books= sorted(list(enumerate(cosine_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_books:
        print(pt.index[i[0]])
```

```python
recommend('A Painted House')
```
